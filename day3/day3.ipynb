## Part One

```python
import pandas as pd
import numpy as np
```

```python
# test_input = ["#1 @ 1,3: 4x4","#2 @ 3,1: 4x4","#3 @ 5,5: 2x2"]

test_df = pd.read_table("input.txt", header=None)
test_df
```

```python
#test_df = pd.DataFrame(test_input, columns=["raw"])
#test_df

test_df.rename(columns={0:"raw"}, inplace=True)
test_df
```

```python
test_df['working'] = test_df['raw'].str.replace(r"[#\ ]","")
test_df[['id','working']] = test_df['working'].str.split("@", expand=True)
test_df[['left_margin','working']] = test_df['working'].str.split(",", expand=True)
test_df[['top_margin','working']] = test_df['working'].str.split(":", expand=True)
test_df[['width','height']] = test_df['working'].str.split("x", expand=True)
test_df.drop('working', inplace=True, axis=1)
test_df
```

```python
test_df['total_width_index'] = test_df.apply(lambda row: int(row['left_margin']) + int(row['width']), axis = 1)
test_df['total_height_index'] = test_df.apply(lambda row: int(row['top_margin']) + int(row['height']), axis = 1)
test_df
```

```python
print("max width: " + str(test_df['total_width_index'].max()))
print("max height: " + str(test_df['total_height_index'].max()))
```

```python
claimed_indices = []
for index, row in test_df.iterrows():
    #print(row)
    left_start = row['left_margin']
    left_end = row['total_width_index']
    top_start = row['top_margin']
    top_end = row['total_height_index']
    #fabric.loc[top_start:top_end, left_start:left_end] = row['id']
    for x in range(int(left_start),int(left_end)):
        for y in range(int(top_start),int(top_end)):
            claimed_indices.append((x,y))
    #print(claimed_indices)
```

```python
from collections import Counter
counted = Counter(claimed_indices)
#print(counted)
multi_claim = 0
for elem in counted:
    if counted[elem] > 1:
        multi_claim +=1
print(multi_claim)
```

```python
test_df['indices'] = [[]] * len(test_df)
#print(test_df)
for index, row in test_df.iterrows():
    #print(row)
    this_row_indices = []
    left_start = row['left_margin']
    left_end = row['total_width_index']
    top_start = row['top_margin']
    top_end = row['total_height_index']
    #fabric.loc[top_start:top_end, left_start:left_end] = row['id']
    for x in range(int(left_start),int(left_end)):
        for y in range(int(top_start),int(top_end)):
            this_row_indices.append((x,y))
    test_df.at[index,'indices'] = this_row_indices
print(test_df)
```

```python
from collections import Counter
counted = Counter(claimed_indices)
#print(counted)
multi_claim = 0
for elem in counted:
    if counted[elem] > 1:
        multi_claim +=1
print(multi_claim)
```
