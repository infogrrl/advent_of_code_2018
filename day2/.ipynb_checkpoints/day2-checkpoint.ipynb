```python
import pandas as pd
from collections import Counter
```

### Part 1

```python
"""
    abcdef contains no letters that appear exactly two or three times.
    bababc contains two a and three b, so it counts for both.
    abbcde contains two b, but no letter appears exactly three times.
    abcccd contains three c, but no letter appears exactly two times.
    aabcdd contains two a and two d, but it only counts once.
    abcdee contains two e.
    ababab contains three a and three b, but it only counts once.
"""

test_data = ['abcdef','bababc','abbcde','abcccd','aabcdd','abcdee','ababab']
```

```python
input_data_df = pd.read_table("input.txt", header=None)
input_data = list(input_data_df[0])
input_data
```

```python
len(input_data)
```

```python
twos = 0
threes = 0

for box_id in test_data:
    letter_dict = Counter(box_id)
    print(letter_dict)

    has_2 = False
    has_3 = False
    for x in letter_dict.values():
        if x == 2:
            has_2 = True
        if x == 3:
            has_3 = True
    if has_2:
        twos += 1
    if has_3: 
        threes += 1
checksum = twos * threes
print(checksum)
```

```python
twos = 0
threes = 0

for box_id in input_data:
    letter_dict = Counter(box_id)
    print(letter_dict)

    has_2 = False
    has_3 = False
    for x in letter_dict.values():
        if x == 2:
            has_2 = True
        if x == 3:
            has_3 = True
    if has_2:
        twos += 1
    if has_3: 
        threes += 1
checksum = twos * threes
print(checksum)
```

### Part 2

```python
import itertools
```

```python
test_input_2 = ['abcde','fghij','klmno','pqrst','fguij','axcye','wvxyz']
```

```python
def get_word_dict(word):
    word_dict = {}
    for index, char in enumerate(word):
        word_dict[index] = char
    return word_dict
```

```python
df = pd.DataFrame(list(itertools.product(test_input_2, test_input_2)), columns = ['id_a','id_b'])
df = df[df['id_a'] != df['id_b']].reset_index(drop=True)
df
```

```python
df['word_dict_a'] = df['id_a'].apply(lambda word: get_word_dict(word))
df['word_dict_b'] = df['id_b'].apply(lambda word: get_word_dict(word))
df
```

```python
def compare_dicts(a, b):
    i = 0
    same = 0
    while i < len(a):
        if a[i] == b[i]:
            same += 1
        i += 1
    if same == (len(a) - 1):
        print(a)
        print(b)
    return same
```

```python
df['dict_compare'] = df.apply(lambda row: compare_dicts(row.word_dict_a,row.word_dict_b), axis=1)
df
```
